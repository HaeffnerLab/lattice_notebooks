```python
s = "scan_data"
d100 = ["1532_54", "1543_46", "1548_21", "1552_55"]
d1000 = ["1556_28", "1626_07"]
d100_0 = h.File("RabiFlopping/{}.h5".format(d100[0]), "r")[s]
d100_1 = h.File("RabiFlopping/{}.h5".format(d100[1]), "r")[s]
d100_2 = h.File("RabiFlopping/{}.h5".format(d100[2]), "r")[s]
d100_3 = h.File("RabiFlopping/{}.h5".format(d100[3]), "r")[s]
d1000_0 = h.File("RabiFlopping/{}.h5".format(d1000[0]), "r")[s]
d1000_1 = h.File("RabiFlopping/{}.h5".format(d1000[1]), "r")[s]
y,x = list(d100_0.keys()); list(d100_0.keys())
```

```python
μ, σ = np.nanmean(d100_0[y][()]), np.nanstd(d100_0[y][()])
plt.plot(range(len(d100_0[y][()])), d100_0[y][()], label="{}, std=({:.2e},{:.2e})".format(d100[0], np.sqrt(μ * (1-μ)/ 100), σ))
μ, σ = np.nanmean(d100_1[y][()]), np.nanstd(d100_1[y][()])
plt.plot(range(len(d100_1[y][()])), d100_1[y][()], label="{}, std=({:.2e},{:.2e})".format(d100[1], np.sqrt(μ * (1-μ)/ 100), σ))
μ, σ = np.nanmean(d100_2[y][()]), np.nanstd(d100_2[y][()])
plt.plot(range(len(d100_2[y][()])), d100_2[y][()], label="{}, std=({:.2e},{:.2e})".format(d100[2], np.sqrt(μ * (1-μ)/ 100), σ))
μ, σ = np.nanmean(d100_3[y][()]), np.nanstd(d100_3[y][()])
plt.plot(range(len(d100_3[y][()])), d100_3[y][()], label="{}, std=({:.2e},{:.2e})".format(d100[3], np.sqrt(μ * (1-μ)/ 100), σ))
plt.ylim(0, 1)
plt.legend(fontsize=15, bbox_to_anchor=(1,1), title="time, std=(theory, exp)", title_fontsize=20)
plt.ylabel("Excitation", fontsize=20)
plt.xlabel("Iteration", fontsize=20)
plt.title("100 Reps", fontsize=20)
```

```python
μ, σ = np.nanmean(d1000_0[y][()]), np.nanstd(d1000_0[y][()])
plt.plot(range(len(d1000_0[y][()])), d1000_0[y][()], label="{}, std=({:.2e},{:.2e})".format(d1000[0], np.sqrt(μ * (1-μ)/ 1000), σ))
μ, σ = np.nanmean(d1000_1[y][()]), np.nanstd(d1000_1[y][()])
plt.plot(range(len(d1000_1[y][()])), d1000_1[y][()], label="{}, std=({:.2e},{:.2e})".format(d1000[1], np.sqrt(μ * (1-μ)/ 1000), σ))
plt.legend(fontsize=15, bbox_to_anchor=(1,1), title="time, std=(theory, exp)", title_fontsize=20)
plt.ylim(0, 1)
plt.ylabel("Excitation", fontsize=20)
plt.xlabel("Iteration", fontsize=20)
plt.title("1000 Reps", fontsize=20)
```
