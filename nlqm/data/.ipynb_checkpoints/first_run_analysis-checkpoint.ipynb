```python
s = "scan_data"
kb = 1.38064852e-23
γ397 = 1 / 7e-9
ħ = 6.62607004e-34 / (2 * np.pi)
ω = 2 * np.pi * 1.02e6
η = 2 * np.pi * np.sqrt( ħ / (2 * ω * 6.6551079e-26)) / 729e-9; η
```

```python
def thermal_carrier(time, Ω, n̄, t0=0):
    A =1
    res = [A/2 for t in time]
    N = 200
    for n in range(0, N):
        res -= A * (n̄ / (n̄ + 1))**n * np.cos(2 * np.pi * Ω * (1 - η**2 * n) * (time - t0)) / (2 * (n̄ + 1))
    return res
```

## Blue sideband Rabi (for temperature estimate + pi-time)

```python
h = h5.File("2115_33.h5", "r")[s]
X = h[xb][()] * 1e6
Y = h[yb][()]
```

```python
def laguerre(x, n):
    L = 1.0, -x + 2
    if n < 2:
        return L[n]
    for i in range(2, n+1):
        L = L[1], ((2 * i - x) * L[1] - i * L[0]) / i
    return L[1]
    
def Pn(n̄, n):
    return (n̄ / (n̄ + 1))**n / (n̄ + 1)

def bsbrabi(time, Ω, n̄1, t0):
    n̄2 = 0
    α = 1
    p = []
    for t in time:
        pi = 0
        for n in range(200):
            pi += (α * Pn(n̄1, n) + (1-α) * Pn(n̄2, n)) * np.sin(Ω/2 * np.exp(-η**2/2) * η * np.sqrt(1/(n+1)) * laguerre(η**2, n) * (t-t0))**2
        p.append(pi)
    return p
```

```python
cf = 70
popt, pcov = curve_fit(bsbrabi, X[:cf], Y[:cf], sigma=np.sqrt(Y[:cf] * (1 - Y[:cf])) / 10 + 1e-4, p0=[0.4, 0.04, 1])
yfit = bsbrabi(Xt, *popt)
```

```python
def carrier(time, Ω, γ, t0):
    return (1 - np.exp(-γ*time) * np.cos(2* np.pi * Ω * (time - t0)))/2
popt, pcov = curve_fit(carrier, X, Y, sigma=np.sqrt(Y * (1 - Y)) / 10 + 1e-3, maxfev=10000, p0=[2.51, 1/18, 0.1]); popt
```

```python
dim = 50
h = qeye(dim)

ψ0 = basis(dim, 0)

γ = 2e-5
nh = 1e4
# γ*nh = 2 quanta / 100 ms = 0.02

cops = [np.sqrt(γ * (nh + 1)) * destroy(dim), np.sqrt(γ * nh) * create(dim)]

ttime = np.linspace(0, 100, 1000)
res = mesolve(h, ψ0, ttime, cops, [num(dim)])
```

```python
print(max(res.expect[0]))
plt.plot(res.times, res.expect[0])
```

```python
dim = 50
n̄ = 0

ρe = thermal_dm(dim, n̄)
ψi_ramsey = (basis(2, 0) + basis(2, 1)) / np.sqrt(2)
ψi_flop = basis(2, 1)

# ρi = ket2dm(ψi_flop)
ρi = ket2dm(ψi_ramsey)

ρ0 = tensor(ρi, ρe)

time_bsb = np.linspace(0, 0.24224, 100)
time_free = np.linspace(0, 50, 2000)

H1 = 2 * np.pi * (tensor(sigmap(), create(dim)) + tensor(sigmam(), destroy(dim)))
# H2 = 2 * np.pi * popt[0] * tensor(qeye(2), num(dim))
H2 = 2 * np.pi * 1e-3 * tensor(qeye(2), num(dim))
```

```python
result_bsb1 = mesolve(H1, ρ0, time_bsb, [], [])
fig, ax = plt.subplots()
ax.plot(result_bsb1.times, expect(tensor(ket2dm(basis(2,0)) , qeye(dim)), result_bsb1.states))
ax.set_ylim(0, 1)
```

```python
ρ0 = result_bsb1.states[-1]
result_free = mesolve(H2, ρ0, time_free, [tensor(qeye(2), i) for i in cops], [])
```

```python
Ry = Qobj([[1 / np.sqrt(2), 1/np.sqrt(2)], [-1 / np.sqrt(2), 1/np.sqrt(2)]])
rotated = []
for state in result_free.states:
    result_bsb2 = mesolve(H1, state, time_bsb, [], [])
    rotated.append(Ry * ptrace(result_bsb2.states[-1], 0) * Ry.dag())
```

```python
plt.plot(time_free, expect(ket2dm(basis(2,0)), rotated).real)
# plt.plot(time_free, (1 + np.exp(-0.04 * time_free)) / 2)
plt.ylim(0, 1)
```

```python
Xi_full = np.concatenate((X1, X2+0.1))
Yi_full = np.concatenate((Y1, Y2))

fig, ax = plt.subplots(figsize=(16,5))
ax.errorbar(Xi_full-popt[-1], Yi_full, yerr=np.sqrt(Yi_full * (1 - Yi_full)) / 10, marker="o", color="k", lw=0, elinewidth=1)
ax.plot(time_free, expect(ket2dm(basis(2,0)), rotated).real, label="fit to Markovian heating at rate 20 quanta/second")
ax.vlines(x=11.57, ymin=0, ymax=1, color="C3", ls="dashed", zorder=-1, lw=2, label="fixed interrogation times")
ax.vlines(x=20.36, ymin=0, ymax=1, color="C3", ls="dashed", zorder=-1, lw=2)
ax.set_ylabel("Probability of\ndark state detection", fontsize=20)
ax.set_xlabel("Ramsey wait time (ms)", fontsize=20)
ax.set_xlim(0, 25)
ax.set_ylim(0, 1)
plt.legend(fontsize=20, loc=1, framealpha=1)
plt.tight_layout()
plt.savefig("/Users/josephbroz/Desktop/nlqm/long_ramsey.pdf")
plt.show()
```
