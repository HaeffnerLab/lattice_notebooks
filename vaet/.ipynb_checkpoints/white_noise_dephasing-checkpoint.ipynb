```julia
using QuantumOptics
import PyPlot
const plt = PyPlot
using LaTeXStrings
```

```julia
sb = SpinBasis(1//2)
```

```julia
H(Δ) = 2π * 1/2 * sigmax(sb) + 2π * Δ/2 * sigmaz(sb)
J(Γ) = [sqrt(2π * Γ/2) * sigmaz(sb)]
```

```julia
ρ0 = dm(spindown(sb))
T = [0:1e-2:5;];
```

```julia
fig, ax = plt.subplots()
for i in 0:.1:1
    Γ = i / 2
    _, ρt = timeevolution.master(T, ρ0, H(1), J(Γ))
    ax.plot(T, real.(expect(dm(spinup(sb)), ρt)), label="Γ=$(Γ)Ω")
end
plt.legend(bbox_to_anchor=(1,1))
plt.title(L"$H = \frac{Ω}{2}σ_x + \frac{Δ}{2}σ_y$")
plt.ylabel("Excitation")
plt.xlabel("Time (2π/Ω)")
plt.ylim(0,1)
plt.tight_layout()
plt.savefig("/Users/josephbroz/Desktop/white_noise_dephasing.pdf")
```
