```python
data = {}
def set_data(name, files):
    if type(files) != list:
        files = [files]
    xdata = []
    ydata = []
    for file in files:
        s = "scan_data"
        d = h.File("SingleIonVAET/{}.h5".format(file), "r")[s]
        y, x = list(d.keys())
        xdata += list(d[x][()] * 1e6)
        ydata += list(d[y][()])
    data[name + "-x"] = xdata 
    data[name + "-y"] = ydata 
```
